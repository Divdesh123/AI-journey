## Prerequisites before starting

* Install Langchain using: ``` pip install -U langchain ``` .
* LangChain provides integrations to hundreds of LLMs and thousands of other integrations. These live in independent provider packages. Install any one of these:
```pip install -U langchain-openai ```  or
``` pip install -U langchain-anthropic```
* Setup OpenAPI or Claude account and get an API key.
* Set the API_KEY environment variable in your terminal.
