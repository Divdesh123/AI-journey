## Ch 1 : Machine Learning Systems in Production

* Google’s 2016 leap in Translate showcased deep learning’s power, sparking ML’s rapid adoption across life and highlighting that real ML systems include not just models but data, infrastructure, hardware, and user interfaces.
* Because ML systems are large, powerful, and impactful, deploying them brings serious engineering and social challenges, and this chapter explains when to use ML, real-world use cases, and how to build reliable, scalable production systems.


### When to Use Machine Learning?
* Although powerful and widely adopted, ML is not a magic solution, and you should evaluate whether it’s truly necessary or cost-effective before starting an ML project.
* Machine learning is an approach to (1) learn (2) complex (3) patterns from
(4) existing data and use these patterns to make (5) predictions on (6)
unseen data
* Learn : An ML system must be able to learn from data, unlike traditional databases, using examples to infer relationships and make predictions for new inputs.
* Complex : ML is useful when patterns are too complex to define manually, enabling systems to learn relationships that are simple for humans to perceive but difficult to encode explicitly.
* Pattern : ML works only when meaningful patterns exist to be learned, unlike truly random processes, though determining whether patterns are present often requires rigorous experimentation.
* Existing Data : ML requires data-either existing, collected, or gathered over time-since without data (or online learning), a model cannot learn meaningful patterns.
* Predictions : ML is useful only for problems that can be framed as predictions, often providing fast, approximate answers where exact computation is costly.
* Unseen Data : ML works only when unseen data follows similar patterns as the training data, allowing learned relationships to generalize effectively.
